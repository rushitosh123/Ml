```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\overall_practice_Files\\Csv_files\\placement.csv")
```

```python
df1 = pd.read_csv("D:\\overall_practice_Files\\Csv_files\\placements.csv")
```

```python
df1
```

```python
plt.scatter(df1['cgpa'],df1['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
plt.show()
```

```python
X = df1.iloc[:,0:1]
y = df1.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split  # Corrected import statement

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)  # Corrected function syntax
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr
```

```python
lr.fit(X_train, y_train)
```

```python
plt.scatter(df1['cgpa'],df1['package'])
plt.plot(X_train,lr.predict(X_train), color="red")
plt.xlabel = ("CGPA")
plt.ylabel = ("Package (in LPA)")
              
```

```python
# now find the accuracy by using differnt evaluation metrics
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 2 cheze batao yaha pe y_test or y_pred means testing data me prediction kya aya
```

```python
y_pred = lr.predict(X_test)
```

```python
y_test
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
# 2.8 lpa average salary 
```

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# 3.4 lpa average salary 
```

```python
# now  here it show how exact our model performance 
print("R2Score",r2_score(y_test,y_pred))
```

```python
# accuracy is 78% variance in lpa 
```

```python
r2= r2_score(y_test,y_pred)
```

```python
# Adjusted R2 Score 
X_test.shape
```

```python
adjustedRscore =  1-((1-r2)*(40-1)/(40-1-1))
```
