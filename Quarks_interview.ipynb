```python
# Write a program to print all the Leaders in the array. 
# An element is a Leader if it is greater than all the elements to its right side.
```

```python
def find_leaders(arr):
    arr= list(arr)
    leaders=[]
    n= len(arr)
    print("qqq",arr)
    
    max_from_right=arr[-1]
    print("11111111",max_from_right)
    leaders.append(max_from_right)

    for i in range(n-2, -1, -1):
        if arr[i] > max_from_right:
            max_from_right = arr[i]
            leaders.append(max_from_right)
    leaders.reverse()
    return leaders

Input1 = {16, 17, 4, 3, 5, 2}
Input2 = {1, 2, 3, 4, 5, 2}

res=find_leaders(Input1)
result= find_leaders(Input2)

print(res)
print('*'*50)
print(result)
```

```python
def find_leaders(arr):
    arr= sorted(arr, reverse=True)
    leaders=[]
    n= len(arr)
    
    max_from_right=arr[0]
    leaders.append(max_from_right)

    for i in range(1,n):
        if arr[i] < max_from_right:
            max_from_right = arr[i]
            leaders.append(max_from_right)
    return leaders

Input1 = {16, 17, 4, 3, 5, 2}
Input2 = {1, 2, 3, 4, 5, 2}

res=find_leaders(Input1)
result= find_leaders(Input2)

print(res)
print('*'*50)
print(result)
```
